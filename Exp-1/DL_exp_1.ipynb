```python
# NumPy
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])


print(a + b)
print(a * b)
print(a ** 2)


# PyTorch
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])


print(a + b)
print(a * b)
print(a ** 2)
```

```python
# NumPy
A = np.array([[10,20,30],[40,50,60]])
print(A[0, 1]) # 20
print(A[:, 1]) # column

# PyTorch
T = torch.tensor([[10,20,30],[40,50,60]])
print(T[0, 1])
print(T[:, 1])
```

```python
# NumPy
x = np.array([1, 2, 3, 4, 5])
mask = x > 3
print(x[mask])

# PyTorch
x = torch.tensor([1, 2, 3, 4, 5])
mask = x > 3
print(x[mask])
```

```python
t = torch.arange(12)
print("Original t:", t)
print("Original shape:", t.shape)


# view (only works if memory is contiguous)
t_view = t.view(3, 4)
print("After view(3,4):")
print(t_view)
print("Shape:", t_view.shape)


# reshape (safe version)
t_reshape = t.reshape(3, 4)
print("After reshape(3,4):")
print(t_reshape)
print("Shape:", t_reshape.shape)


# unsqueeze (add dimension)
t_unsq = t.unsqueeze(0) # shape: (1, 12)
print("After unsqueeze(0):")
print(t_unsq)
print("Shape:", t_unsq.shape)


# squeeze (remove dim size=1)
t_sq = t_unsq.squeeze() # shape: (12,)
print("After squeeze():")
print(t_sq)
print("Shape:", t_sq.shape)
```
